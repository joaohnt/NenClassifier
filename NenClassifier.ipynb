```python
from sklearn.tree import DecisionTreeClassifier
import pickle
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
from sklearn.ensemble import RandomForestClassifier
```

# =-=-=-=-=-=-=-=-=-= NenClassifier =-=-=-=-=-=-=-=-=-=

```python
with open('nen_data.pkl', 'rb') as f:             # carregando o arquivo .pkl
    X_nen_treinamento, y_nen_treinamento, X_nen_teste, y_nen_teste = pickle.load(f)
```

```python
X_nen_treinamento
```

```python
y_nen_teste
```

```python
arvore_nen = DecisionTreeClassifier(criterion= 'entropy', random_state = 0)
arvore_nen.fit(X_nen_treinamento, y_nen_treinamento)
```

```python
previsoes = arvore_nen.predict(X_nen_teste)
previsoes
```

```python
y_nen_teste
```

```python
accuracy_score(y_nen_teste, previsoes)
```

```python
previsores = ['personalidade', 'hobbie', 'reacao']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (30, 30))
tree.plot_tree(arvore_nen, feature_names = previsores, class_names = arvore_nen.classes_, filled = True);
fig.savefig('arvore_nen.png')
```

```python
# extrovertido, aventura, impulsivo == [4,3,0]
previsoes = arvore_nen.predict([[4,3,0]])
previsoes
```
